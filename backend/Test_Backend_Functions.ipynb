## Generate Team Rank Model
Create team rank model using the bayes_TeamRankModel function.

```python
beta_Off_cols = list(beta_Off_DF.columns.values)
beta_Off_DF.columns = [x+'_off' for x in beta_Off_cols]

beta_Def_cols = list(beta_Def_DF.columns.values)
beta_Def_DF.columns = [x+'_def' for x in beta_Def_cols]

beta_concat = pd.concat([roboSpecStats,\
                         beta_Off_DF,\
                         beta_Def_DF],\
                        axis=1,\
                        sort=False)
beta_concat = pd.concat([beta_Off_DF,\
                         beta_Def_DF],\
                        axis=1,\
                        sort=False)
```

```python
beta_concat.info()
beta_concat.head(100)
```

```python
fitSampling_DF,fitSummary_DF,betaDict = bayes_TeamRankModel(beta_concat,\
                                                            rankTable['rank'],\
                                                            nIter=5000,\
                                                            nChains=4,\
                                                            nJobs=4,\
                                                            priorType='Finnish_horseshoe',\
                                                            stanDictVar={'m0':10,\
                                                                         'slab_scale':3,\
                                                                         'slab_df':25,\
                                                                         'beta_tilde_mu':0,\
                                                                         'beta_tilde_sigma':1,\
                                                                         'lambda_mu':0,\
                                                                         'lambda_sigma':1,\
                                                                         'tau_tilde_mu':0,\
                                                                         'tau_tilde_sigma':1,\
                                                                         'alpha_mu':0,\
                                                                         'alpha_sigma':2})
```

### Testing Normality for Different Features

```python
data = fitSampling_DF['beta[19]']
result = normalityTests(data)

if not (result is None):
    testResult = result[0]
    normBool = result[1]
    
    print(normBool)
    print(testResult)
else:
    print(result)
```

```python
import numpy as np
from scipy import stats
from numpy.random import randn
import plotly.graph_objects as go

data = fitSampling_DF['beta[19]']

qq = stats.probplot(data, dist='norm', sparams=(1))
x = np.array([qq[0][0][0], qq[0][0][-1]])

fig = go.Figure()
fig.add_scatter(x=qq[0][0], y=qq[0][1], mode='markers')
fig.add_scatter(x=x, y=qq[1][1] + qq[1][0]*x, mode='lines')
fig.layout.update(showlegend=False)
fig.show()
```
